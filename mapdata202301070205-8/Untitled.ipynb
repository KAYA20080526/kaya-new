```python
import geopandas as gpd
County_data = gpd.read_file("/Users/ho-yehcheng/Documents/GitHub/mapdata202301070205-8/mapdata202301070205-11/COUNTY_MOI_1090820.shp")
County_data
```

```python
## plot data
Sample_data = County_data.loc[:, ['COUNTYNAME', 'geometry']]
Sample_data.plot()
```

```python
County_data[['COUNTYNAME', 'geometry']]
```

```python
County_data = County_data. rename(columns={ 'geometry': 'county_geometry'})

# 合併'Town_data'和'County_data'避免和後面要合併的'Town_data'中的'geometry'列重名
merged_data = Rice_data.merge(County_data[ ['COUNTYNAME', 'county_geometry']], on='COUNTYNAME', how='left' )

# # 將合併後的'county_geometry’列的值插入'Town_data'的'geometry'列，這樣 merged_data 就有了'geometry'列。
merged_data ['geometry'] = County_data['county_geometry']

# # 顯示合併後的 GeoDataFrame
merged_data
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 將資料轉換為 GeoDataFrame，並將 'geometry' 列重新命名為 'geom'

gdf = gpd.GeoDataFrame(merged_data, geometry='geometry')
gdf.rename_geometry('geom', inplace=True)

# 畫出地圖

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='yield', cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# 添加標題

plt.title('Population Distribution', fontdict={'fontsize': '15', 'fontweight' : '3'})

# 顯示圖表

plt.show()
```
